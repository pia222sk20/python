```python
# 합성곱 연산 - > 이미지의 특징을 추출
# convolution  - CNN
# 28 28 3  너비 높이 깊이  3차원 텐서
# 커널은 일반적으로 3 x 3, 5 x 5
# 5 x 5 이미지  3 x 3 
```

```python
import torch
import torch.nn as nn
```

```python
# 배치크기 x 채널크기 x 높이 x 너비 
inputs = torch.Tensor(1,1,28,28)  # 28x28 크기를 갖는 한개의 흑백이미지
print(f'텐서의 크기 : {inputs.size()}')
```

```python
# 합성곱층과 폴링 선언
conv1 = nn.Conv2d(1,32,3,padding=1)  # 특성맵 stride 1 padding=1     1 32 28 28 
print(conv1)
```

```python
# 2번째 합성곱 층
conv2 = nn.Conv2d(32,64,3,padding=1)
print(conv2)
```

```python
pool = nn.MaxPool2d(2)
print(pool)
```

```python
inputs.size(), conv1
```

```python
# 데이터 통과하기
out = conv1(inputs)
out.size()
```

```python
out = pool(out)
print(out.size())  # 1 32 14 14
```

```python
out = conv2(out)
print(out.size())
```

```python
out = pool(out)
print(out.size())
```

```python
out = out.view(out.size(0),-1)
print(out.size())
```

```python
fc = nn.Linear(3136,10)
out = fc(out)
print(out.size())
```
