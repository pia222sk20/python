```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# # titanic 데이터셋 로딩
tips = sns.load_dataset('tips').rename(columns={'sex':'gender'})
tips
```

```python
tips.query('total_bill < 10')
```

```python
# 여성이면서 팁이 5달러 이상인 고객
tips.query("gender == 'Female' and tip > 5"  )
```

```python
# 주말(fri, sat, sun) 이면서 저녁에 식사인원 3명 인상인 고객
tips.query("(day in ('Fri','Sat','Sun')) and (time == 'Dinner' and size >= 3)"  )
```
