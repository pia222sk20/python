```python
import pandas as pd

# 데이터프레임 변환
df = pd.read_excel('./data/남북한발전전력량.xlsx')

# # 남한, 북한 발전량 합계 데이터만 추출
df_ns = df.iloc[[0, 5], 2:] 

# 행 인덱스 변경
df_ns.index = ['South','North'] 

# 발전량 데이터의 자료형을 문자열에서 정수형으로 변환
df_ns = df_ns.astype(int)                 

# 데이터프레임 출력
df_ns.head()
```

```python
# 자료형 확인
df_ns.info()
```

```python
# 선 그래프 그리기
# df_ns.plot()
df_ns.iloc[1].plot()
print(df_ns)
```

```python
# 행, 열 전치하여 다시 그리기
tdf_ns = df_ns.T
print(tdf_ns.head())
print('\n')
tdf_ns.plot();
```
