```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.pivot(index='year',columns='month',values='passengers')
```

```python
# 연도별 승객수의 변화 lineplot  - sns
import matplotlib.pylab as plt
import seaborn as sns
# sns.lineplot(data=flights,x='year',y='passengers',errorbar=('ci', 95))
# sns.lineplot(data=flights,x='year',y='passengers',hue='month',  errorbar=None)
sns.lineplot(data=flights,x='year',y='passengers')
# sns 시각화는 신뢰구간도같이 표시하는 경우
# 데이터가 변동될수 있는 구간
# 신뢰구간이 넓으면 불확실성이 크다... (데이터 분산이 크거나 샘플수가 작을경우)
# 신뢰구간이 좁으면 안정적... (데이터가 모여있거나 샘플수가 많을경우)
# plt.legend(loc='best')
plt.show()
```

```python
# 년도별 월별 승객의 변화 시각화
plt.figure(figsize=(12,5))
sns.lineplot(data=flights,x='year',y='passengers',hue='month')
```
